```python
from ultralytics import YOLO

# d2 = second dataset (Traffic-and-Road-Signs)
baseline_model = YOLO('./models/d2/baseline_v1d2.pt')                  # baseline, no augmentation
augmented_model = YOLO('./models/d2/best_v1d2_v1aug.pt')       # v1aug, with mosaic/scale/copy_paste

print("Baseline classes:", baseline_model.names)
print("Augmented classes:", augmented_model.names)
```

```python
import os
print(os.path.exists('./models/d2/baseline_v1d2.pt'))
print(os.path.exists('./models/d2/best_v1d2_v1aug.pt'))
```

```python
test_image = './manual_test_images/testing7.png'  # same image for both

print("=== BASELINE ===")
baseline_results = baseline_model.predict(
    source=test_image, imgsz=640, conf=0.5, save=True,
    project='./outputs/predictions', name='compare_baseline'
)
print(f"Detections: {len(baseline_results[0].boxes)}")

print("\n=== AUGMENTED (v1aug) ===")
augmented_results = augmented_model.predict(
    source=test_image, imgsz=640, conf=0.5, save=True,
    project='./outputs/predictions', name='compare_augmented'
)
print(f"Detections: {len(augmented_results[0].boxes)}")
```
